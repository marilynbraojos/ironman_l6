```python
import os
```

```python
for filename in os.listdir("./2024F_Gimgs/"):
    if filename == "labels.txt":
        continue 

    old_path = os.path.join("./2024F_Gimgs/", filename)
    new_filename = '2024G' + filename
    new_path = os.path.join("./2024F_Gimgs/", new_filename)

    # Rename the file
    os.rename(old_path, new_path)
    print(f"Renamed: {filename} → {new_filename}")
```

```python
updated_lines = []

with open("./2024F_Gimgs/labels.txt", 'r') as f:
    lines = f.readlines()
    for line in lines:
        line = line.strip()
        if not line:
            continue
        image_name, label = line.split(',')
        new_line = f"2024G{image_name},{label}"
        updated_lines.append(new_line)

# Write the updated lines back to the file
with open("./2024F_Gimgs/labels.txt", 'w') as f:
    for line in updated_lines:
        f.write(line + '\n')

print("✅ labels.txt updated.")
```
